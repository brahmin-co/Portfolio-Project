```python
#Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjust the configuration of the plot

pd.options.mode.chained_assignment = None

# Read in the data
df = pd.read_csv(r'C:\Users\hp\OneDrive\Documents\Pandas Project\movies.csv')
```

```python
# take a look at the data
df.head()
```

```python
# See if there is missing Data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# data types for columns
print(df.dtypes)
```

```python
#change data type of columns

df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
df
```

```python
# Create correct year column
df['yearcorrect'] = df['released'].astype(str).str[:4]
```

```python
df
```

```python
# Order our Data a little bit to see

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Scatter plot with budget vs gross


plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
# Plot Budget vs Gross with seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
```

```python
# Find the Correlation

df.corr(numeric_only=True)
```

```python
correlation_matrix = df.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot= True)

plt.title('Correlation Matrix For Numeric Features ')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Look at Company

df.head()
```

```python
df_numerized = df.copy()  # Create a copy of the original dataframe

for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype == 'object':
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
```

```python
correlation_matrix = df_numerized.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot= True)

plt.title('Correlation Matrix For Numeric Features ')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
# Set to display all rows (or a very large number)
pd.set_option('display.max_rows', None)

# Set to display all columns (if necessary)
pd.set_option('display.max_columns', None)

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Filter only those with high absolute correlation values (e.g., > 0.5 or < -0.5)
significant_corr = corr_pairs[(corr_pairs.abs() > 0.5) & (corr_pairs.abs() < 1)]
print(significant_corr)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort", ascending=False)
print(sorted_pairs)
```
